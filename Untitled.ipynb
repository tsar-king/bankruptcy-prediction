```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('data/raw/data.csv')
print(f"Shape: {df.shape}")
print(f"\nColumns:\n{df.columns.tolist()}")
print(f"\nFirst 5 rows:")
print(df.head())
print(f"\nData types:\n{df.dtypes.value_counts()}")
print(f"\nMissing values:\n{df.isnull().sum().sum()}")
print(f"\nTarget distribution:\n{df['Bankrupt?'].value_counts()}")
print(f"\nTarget percentage:\n{df['Bankrupt?'].value_counts(normalize=True) * 100}")
```

```python
# Statistical summary of features
df.describe().T.head(20)
```

```python
# Sample a subset for correlation plot (95 features too many to plot all)
feature_cols = [c for c in df.columns if c != 'Bankrupt?']
corr_matrix = df[feature_cols].corr().abs()

# Find pairs with correlation > 0.9
high_corr = []
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if corr_matrix.iloc[i, j] > 0.9:
            high_corr.append((corr_matrix.columns[i], corr_matrix.columns[j], corr_matrix.iloc[i, j]))

print(f"Number of feature pairs with correlation > 0.9: {len(high_corr)}")
print("Top 10 highly correlated pairs:")
for pair in high_corr[:10]:
    print(f"  {pair[0]} - {pair[1]}: {pair[2]:.3f}")
```
